```python
# remove average

emp_df = emp_sur_df.drop([9])

stats_df = emp_df.describe()
r_stats_df = stats_df.round(1)
r_stats_df
```

```python
avg_stats = r_stats_df.loc[["mean"]]
avg_stats
```

```python
# transpose rows and columns
stats_hbar = avg_stats.T
sorted_avg = stats_hbar.sort_values("mean")
sorted_avg.reset_index()
```

```python
sorted_avg.plot.barh(x=None, y="mean", figsize=(20, 30))
```
